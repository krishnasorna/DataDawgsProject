```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ncaa_31_guards_2021 = pd.read_csv('NCAA_31_Guards_2021.csv').rename(columns={'3 Points Made': 'Three_Points_Made','3PA':'Three_PA','FG%':'FG_Percent','3P%':'Three_P_Percent','FT%':'FT_Percent'})
nba_31_guards_2021 = pd.read_csv('NBA_31_Guards_2021.csv').rename(columns={'3 Points Made': 'Three_Points_Made','3PA':'Three_PA','FG%':'FG_Percent','3P%':'Three_P_Percent','FT%':'FT_Percent'})
```

```python
ncaa_31_guards_2021.columns
```

```python
#copying the ncaa 31 guards 2021 data and adding a response column of PTS_NBA
train1 = ncaa_31_guards_2021.copy()
train1['PTS_NBA'] = nba_31_guards_2021.PTS

#Multiple linear regression with all possible predictors, no interactions or transformations
ols_object1 = smf.ols(formula = 'PTS_NBA~GP+GS+MIN+PTS+FGM+FGA+FG_Percent+Three_Points_Made+Three_PA+Three_P_Percent+FTM+FTA+FT_Percent+OFF+DEF+TRB+AST+STL+BLK+TOV+PF', data = train1)
model1 = ols_object1.fit()
model1.summary()
```

```python
#calculating VIFs for each of the predictors in the MLR w/ all possible variables
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant
X = train1[['Player', 'GP', 'GS', 'MIN', 'PTS', 'FGM', 'FGA', 'FG_Percent',
       'Three_Points_Made', 'Three_PA', 'Three_P_Percent', 'FTM', 'FTA',
       'FT_Percent', 'OFF', 'DEF', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF']]
X = add_constant(X)
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns

for i in range(len(X.columns)):
    vif_data.loc[i,'VIF'] = variance_inflation_factor(X.values, i)

print(vif_data)
```
